```python
import pandas as pd
df = pd.read_csv("https://api.covid19india.org/csv/latest/states.csv")
df
```

```python
df = df[df['Date']!='2021-04-16']
```

```python
df
```

```python
confirmed_india = []
dates_india = []
for i in range(len(df)):
  if(df.iloc[i]['State']=="India"):
    confirmed_india.append(df.iloc[i]['Confirmed'])
    dates_india.append(df.iloc[i]['Date'])
for i in range(len(confirmed_india)):
  print(dates_india[i],confirmed_india[i])
```

```python
states = {}
for i in range(len(df)):
  states[df.iloc[i]['State']] = 1
```

```python
import matplotlib.pyplot as plt
```

```python
import datetime
for state in states:
  dates = []
  confirmed_cases = []
  for i in range(len(df)):
    if df.iloc[i]['State'] == state:
      t = df.iloc[i]['Date'].split('-')
      v = datetime.datetime(int(t[0]),int(t[1]),int(t[2]))
      dates.append(v)
      confirmed_cases.append(df.iloc[i]['Confirmed'])
  plt.figure(figsize=(20,10))
  plt.plot(dates,confirmed_cases)
  stri = "Confirmed cases in "+state
  plt.title(stri)
  plt.xlabel("Dates")
  plt.ylabel("Confirmed cases")
  print()
    
```

```python
india_dates = []
india_confirmed = []
for i in range(len(df)):
  if df.iloc[i]['State'] == "India":
    t = df.iloc[i]['Date'].split('-')
    v = datetime.datetime(int(t[0]),int(t[1]),int(t[2]))
    india_dates.append(v)
    india_confirmed.append(df.iloc[i]['Confirmed'])
plt.plot(india_dates,india_confirmed)
```

```python
for i in range(len(india_dates)):
  print(india_dates[i],india_confirmed[i])
```

```python
new_cases = []
new_cases.append(india_confirmed[0])
for i in range(1,len(india_dates)):
  new_cases.append(india_confirmed[i]-india_confirmed[i-1])
plt.figure(figsize=(19,8))
plt.ylabel("New Cases")
plt.xlabel("Date")
plt.plot(india_dates,new_cases)
plt.show()
```
